```python
def comparativo_shock(data, erro_lab,titulo, plot_diff=True, print_output=True,
                   size_text=11, size=(8,15), xlim=[0, 0.16], ylim=[-15, 30]):
    """
    Plota comparativo dos dados de vibraçao (senoidal e choque).
    -----
    data = dataframe com os dados dos dois ensaios
    erro_lab = erro laboratorial em decimal Ex.: 0.10 (10%)
    plot_diff = condicao para plotar ou nao o grafico com o erro (True ou False)
    print_output = condicao para mostrar os dados do gmrs total e o valor do erro normalizado no grafico (True ou False)
    size_text = tamanho da fonte 
    size = tamanho do grafico
    -----
    """
    
    columns = data.columns # adquire os nomes das colunas
    erro = (data[columns[0]] - data[columns[1]]) # calcula o erro entre a coluna 1 e a coluna 2
    
    max_ind = erro.abs().idxmax() # iindice do maximo do erro (segundos)
    max_value = erro.loc[max_ind] # valor maximo do erro
    
    fig = plt.figure(dpi=100) 
    
    if plot_diff == False:
        fig.set_figheight(size[1]*0.95) # Altura da imagem
    else:    
        fig.set_figheight(size[0]) # Altura da imagem
    
    fig.set_figwidth(size[1]) # Largura da imagem
    # picos dos ensaios
    max_id1_sho = data[columns[0]].abs().idxmax() # indice do pico do 1o ensaio
    max_id2_sho = data[columns[1]].abs().idxmax() # indice do pico do 2o ensaio
    max_id_sho = [max_id1_sho, max_id2_sho] 
    
    max_value1_sho = data[columns[0]].loc[max_id1_sho] # valor maximo do 1o ensaio
    max_value2_sho = data[columns[1]].loc[max_id2_sho] # valor maximo do 2o ensaio
    max_values_sho = [max_value1_sho, max_value2_sho]
    
    # calculo do erro normalizado dos ensaios de choque e criacao do dataframe
    output = pd.Series(columns, name='Acelerometro')
    output = pd.DataFrame(output)
    output['Max. Freq.[Hz]'] = np.array(max_id_sho)
    output['Max. Pico'] = np.array(max_values_sho)
    output['Incerteza Expand.'] = np.array(max_values_sho) * erro_lab
    
    En_sho = abs(output['Max. Pico'].diff()[1]) / np.sqrt((output['Incerteza Expand.']**2).sum())
    print(f'{round(output,3)}\n\nErro Nomalizado: {round(En_sho,3)}')
            
    ax1 = plt.subplot2grid(shape=(3, 3), loc=(0, 0), colspan=3, rowspan=2)
    ax1.scatter(max_id_sho, max_values_sho, c='magenta', linewidths=1)
    ax1.plot(data)
    ax1.set_xlim(xlim)
    ax1.set_ylim(ylim)
    #ax1.set_xscale('log')
    #ax1.set_yscale('log')
    ax1.legend(data.columns)
    ax1.set_ylabel('g (0-pk)')
    ax1.title.set_text(titulo)
    # condicao para plotar o dataframe com os valores maximos e o erro normalizado
    if print_output == True:
        ax1.annotate(f'{round(output,6)}\n\nErro Nomalizado: {round(En_sho,3)}',
                     ((xlim[0]+0.001)*1.2, ylim[0]*0.95), size=size_text); # horizontal alignment can be left, right or center
    # condicao para plotar o grafico com o erro entre os ensaios
    if plot_diff == True:
        ax2 = plt.subplot2grid(shape=(3, 3), loc=(2, 0), colspan=3, rowspan=1) 
        ax2.plot(erro, '--', label=erro.name, linewidth=1) # Dados principais Interpolados
        ax2.axvline(x=max_ind, c='magenta', alpha=0.3, zorder=1) # data do ponto maximo
        ax2.scatter(max_ind, max_value, c='magenta', linewidths=1)
        ax2.set_xlim(xlim)
        #ax2.set_xscale('log')
        # anotacao do valor maximo e seu respectivo tempo no grafico do erro entre os ensaios
        ax2.annotate(f'Max value:{round(max_value,6)} g \n@ Time: {max_ind} s', # this is the text
                         (max_ind, max_value), # this is the point label
                         textcoords='offset points', # how to position the text
                         xytext=(10,0), # distance from text to points (x,y)
                         ha='left',
                         size=size_text); # horizontal alignment can be left, right or center
        ax2.set_xlabel('Time [s]')
        ax2.set_ylabel('Erro [g]');
        
    try:
        os.mkdir('./img/shock')
    except:
        pass
    
    plt.savefig(f'./img/shock/{data.columns[0]} x {data.columns[1]} (sem tabela erro normalizado)', dpi=100, bbox_inches='tight')
    
    #return (output, En)
    
    return (output, En_sho)
```

```python
def time_pulse(df, filtro, display_df=True):
    if display_df == True:
        _ = df.loc[filtro[0]:filtro[1]]
        #df_[df_ >= 0].dropna(how='all')
        print(_)
        _.plot()
        plt.show()
        print()
        print('='*150)
        print()
        
    for i in df.columns:
        df_temp = df[i].loc[filtro[0]:filtro[1]]
        df_temp.plot(label=i, style='.')
        #plt.hlines(0, xmin=df_temp.index[0], xmax=df_temp.index[-1], colors='black', linewidths=0.5) 
        
        df_temp = df_temp[df_temp >= 0].dropna().reset_index()   
        first = df_temp.iloc[0]['TIME']
        last = df_temp.iloc[-1]['TIME']
        delta =  round((last - first) * 1000 , 2)
        print(f'Inicio pulso: @{first}s - Final do pulso: @{last}s')
        print(f'delta t: {delta} ms')
        
        plt.vlines(first, ymin=-10, ymax=30, colors='magenta', alpha=0.5)
        plt.vlines(last, ymin=-10, ymax=30, colors='magenta', alpha=0.5)
        plt.legend()
        plt.show()
```

```python
df = df_shock[['ER1ZSH03_P1-CTRL', 'ER2ZSH03_P1-CTRL']]
time_pulse(df, filtro=[0.109570, 0.12014], display_df=False)
```
